# Linear Regresion

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

sns.set_style("darkgrid")
```

# Create data

```python
# Make x and y have the same space 0 to 100 and generate 50 points
y = np.linspace(0, 100, 50)

# Add randomness, notice scale refers to std dev.
var1 = np.linspace(0, 100, 50) + np.random.normal(scale=20, size=50)

var2 = np.linspace(0, 100, 50) + np.random.normal(scale=10, size=50)
```

# Build the DataFrame

```python
df = pd.DataFrame({'y':y, 'var1':var1, 'var2':var2})
```

# Plot linear model

```python
sns.lmplot(x='var1', y='y', data=df)

sns.lmplot(x='var2', y='y', data=df)
```

# Linear Regression

## Create train, test datasets

```python
# sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=40)
```

```python
y_test.shape
```

## Fit the model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## Predict

```python
y_predict = model.predict(x_test)
y_predict
```

## Score

```python
model.score(x_test,y_test)
```

## Coefficients and Intercept

```python
model.coef_
```

```python
model.intercept_
```
