```python
# From Gemini responding to the question: "what is an iterator in python?"
# An iterator contains two special methods: __iter__ and a __next__ 
# Modified Gemini response by simply adding some comments within the code (OAT 4/11/2024)

class FibonacciIterator:
    def __init__(self, max_value):
        """ 
        Creates a Fibonacci sequence.
        The maximum Fibonacci number generated in the iterator
        will be smaller than or equal to max_value.
        
        Input:
        max_value: Numeric.
        """

        self.a, self.b = 0, 1
        self.max_value = max_value

    def __iter__(self):
        """ 
        An iterator requires an __iter__ method.
        A for loop calls __iter__ once to initiate the process.
        
        Returns:
        The object itself to be used in a for loop
        """
        return self

    def __next__(self):
        """        
        An iterator requires a __next__ method.
        It is called repeatedly by a for loop to retrieve the next element.
        It raises StopIteration when it runs out of element to retrieve.
        
        Returns:
        value: Integer, iteration, next Fibonacci value

        """
        if self.a <= self.max_value:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            return value
        else:
            raise StopIteration
```

```python
# The point of this code, from Gemini was to demonstrate how an iterator works.
fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num)
```

```python
# The point of this code, from Gemini was to demonstrate how an iterator works.
fib_iterator = FibonacciIterator(30)
for num in fib_iterator:
    print(num)
```
