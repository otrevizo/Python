```python
def test_digit(x):
    try:
        # If s is a valid digit then no error from casting
        digit = float(x)
        print("This is a digit")
        return True
    except ValueError:
        print("This is not a digit")
        return False
    except TypeError:
        print("Wrong data type")
        return False
    else:
        print("There is some other exception or error")
        return False
```

```python
x = '3.14'
test_digit(x)
```

```python
x = 3.14
test_digit(x)
```

```python
x = 3
test_digit(x)
```

```python
x = 'abc'
test_digit(x)
```

```python
x = True
test_digit(x)
```

```python
x = '1.2.3'
test_digit(x)
```

```python
x = None
test_digit(x)
```
