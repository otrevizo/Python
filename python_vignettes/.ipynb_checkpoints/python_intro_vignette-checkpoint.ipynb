# Variables 
variables are name representing value stored in the memory.

## Variable Naming rule
I. you cannot use python's keywords for naming variable examples:

1. and 
2. del 
3. from 
4. None 
5. True
6. as 
7. elif
8. def
9. else   
are some examples.

II. No spaces between variable names. For example the variable " fruit name" is incorrect

III. first character a-z, A-Z or underscore .Examples : a , _a, fruit, Fruit 

IV. After first character you can use a-z, A-Z , 0-9 or underscore .Examples : a20, a_20, _20 

V. Case sensitive ,  fruit, and Fruits are considered distinct.


```python
# Declaring and storing values in variables
fruit1_count = 10        # int
fruit1_price = 0.75      # float
fruit1_name = 'apple'    # string

print(fruit1_name, fruit1_count, fruit1_price)
```

# if statement

Used to create a decision structure. Executes one or more statements only if a boolean condition is true.

<table>
    <tr><th>Operator</th><th>Description</th></tr>
    <tr><td>&gt;</td><td>greater than</td></tr>
    <tr><td>&lt;</td><td>less than</td></tr>
    <tr><td>&gt;=</td><td>greater than or equal to</td></tr>
    <tr><td>&lt;=</td><td>less than or equal to</td></tr>
    <tr><td>==</td><td>equal to</td></tr>
    <tr><td>!=</td><td>not equal to</td></tr>
</table>

```python
sales = 10000
bonus = 500
name = 'Ada'
added_bonus = 0

if sales > 9000:
    added_bonus = bonus

print(name + ' total bonus =' + str(added_bonus))
```

## if-else statement
Executes a block of statements if condition is true, or another block if condition is false.

```python
sales = 1000
if sales > 9000:
    added_bonus = 500
    print('You met your sales quota, your bonus = ' + str(added_bonus))
else:
    print('You did not meet the sales quota for bonus')
```

```python
# function with variable number of arguments
def ftn_xargs_sum(*args):
    result = 0
    for x in args:
        result = result + x
    return result
```

```python
ftn_xargs_sum(1, 2, 3)
```

```python
# This function counts the number of digits in the string 
def count_digits(st):
    digit_count = 0
    for i in range(len(st)):
        if st[i].isdigit():
            digit_count = digit_count + 1
    return digit_count
```

```python
count_digits('app43le45')
```

```python
value = 'app43e45'
for i in range(len(value)):
    x = (value[i])
    if x.isdigit() == True:
        print("Digit")
    else:
        print("Non Digit")
```
