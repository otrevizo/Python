## Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_style("darkgrid")
```

# Create data

```python
# Make x, y, and z have the same space 0 to 100 and generate 50 points

# Start with x making it 0 to 100 points
x = np.linspace(0, 100, 50) + np.random.normal(scale=15, size=50)

# Now y make it a positive 0 to 100
y = np.linspace(0, 100, 50) + np.random.normal(scale=20, size=50)

# Make it negative slope for z
z = np.linspace(100, 0, 50) + np.random.normal(scale=10, size=50)
```

# Build the DataFrame

```python
df = pd.DataFrame({'x':x, 'y':y, 'z':z})
```

# Plot it

```python
sns.lmplot(x='x', y='y', data=df)

sns.lmplot(x='z', y='y', data=df)
```

# Covariance

Covariance for population:

$$ cov(x,y) = \frac{{\sum_{i=1}^{n}{(x_i-\mu_x)*(y_i-\mu_y)}}}{n} $$

Covariance for sample:

$$ cov(x,y) = \frac{{\sum_{i=1}^{n}{(x_i-\mu_x)*(y_i-\mu_y)}}}{n-1} $$


```python
df.cov()
```

# Pearson Correlation

Pearson's coefficient provide a number betwwen -1 and 1 (a type of score going from negative to positive correlation).

$$ \rho = \frac{cov(x,y)}{\sigma_x\sigma_y} $$

```python
df.corr()
```
