## _Python Object Oriented (OO)_
### _class WppPipeline_
**United Nations and World Bank**

The United Nations publishes demographics and economics data that includes 
data on migration, country-by-country and region by region. The World Bank 
provides additional economics data, including GDP growth per country, per 
region and worldwide that can be used as exogenous variables in time series 
analysis of migration.

The `WppPipeline` class encapsulates the full data pipeline — loading, 
cleaning, merging, normalizing, and building the time series DataFrame. 
Once instantiated and built, the pipeline object holds all the data 
needed for downstream analysis and visualization.

**Methods:**
- `load_un()` — loads and cleans the UN World Population Prospects Excel file
- `load_wb()` — loads and cleans the four World Bank GDP and inflation files
- `merge()` — merges UN and World Bank data on Location and Year, keeps countries only
- `remove_outliers()` — removes extreme migration rate outliers and low population countries
- `normalize()` — normalizes selected numeric columns with `N_` prefix
- `build_time_series()` — builds normalized time series via year-by-year OLS regression
- `to_csv()` — writes three CSV files for record keeping and reproducibility
- `build()` — convenience method that runs the full pipeline in sequence

**Data available after `pipeline.build()`:**
- `pipeline.df` — merged, cleaned DataFrame (UN + World Bank, countries only)
- `pipeline.n_df` — normalized version of `pipeline.df`
- `pipeline.ts_df` — normalized time series indexed by Year, ready for `TimeSeries` class
- `pipeline.outlier_countries` — list of countries removed by z-score threshold
- `pipeline.low_pop_countries` — list of countries removed by population threshold

**Usage:**
```python
pipeline = WppPipeline(
    un_file            = '../data/WPP2024_GEN_F01_DEMOGRAPHIC_INDICATORS_COMPACT_20260327.xlsx',

    wb_inflation_file  = '../data/WB_inflation.xls',
    year_start         = 1961,
    year_end           = 2023
)
pipeline.build()
pipeline.to_csv()
```


#### _United Nations World Population Prospects_

The United Nations World Population Prospects dataset gives us insights from net migration and migration rates, to explore trends over time. The data is suitable to generate time series, visualization, and serial correlation. The time series analysis considers time periods that may be characterized by world events or regime changes. We will explore possible cycles.

The dataset provides records for a specific country or from a group of countries with various classifications. In the dataset provided by the UN, the group of countries labeled under 'Central America' includes Mexico. Normally Mexico is under North America. Here, the UN classified Mexico under Central America. To avoid confusion with the classic geographic classification for North America (North America includes Mexico, the US, and Canada), the report used the word 'Northern' instead 'North' in 'Northern America' to include only Canada and the US under 'Northern America'.

https://population.un.org/wpp/Download/SpecialAggregates/EconomicTrading/
http://data.un.org/_Docs/SYB/CSV/SYB65_230_202209_GDP%20and%20GDP%20Per%20Capita.csv

#### _World Bank GDP and Inflation Indicators_

The World Bank - World Development Indicators - publishes a data market that
offers several views of GDP and inflation worldwide from 1960 onward.
The goal here is to merge it with the demographic indicators from the United
Nations. Some of these columns will serve as exogenous variables to understand
the behavior of worldwide migration before we focus our attention on the
United States and Mexico.

Four indicators are used, downloaded from the World Bank data portal.
The World Bank uses a standard indicator code naming convention:
`NY` (National Accounts), `GDP` (Gross Domestic Product),
`MKTP` (Market Price), `PCAP` (Per Capita), `KD` (Constant dollars),
`ZG` (Growth rate), `DEFL` (Deflator/Inflation).

Data sources (accessed March 30, 2026):

| Simple filename | WB Indicator code | Description | Original downloaded filename |
|---|---|---|---|
| `WB_GDP.xls` | `NY.GDP.MKTP.KD` | GDP in constant US$ | `API_NY.GDP.MKTP.KD_DS2_en_excel_v2_4.xls` |
| `WB_GDP_growth.xls` | `NY.GDP.MKTP.KD.ZG` | GDP growth (%) | `API_NY.GDP.MKTP.KD.ZG_DS2_en_excel_v2_29.xls` |
| `WB_GDP_pcap_growth.xls` | `NY.GDP.PCAP.KD.ZG` | GDP per capita growth (%) | `API_NY.GDP.PCAP.KD.ZG_DS2_en_excel_v2_227.xls` |
| `WB_inflation.xls` | `NY.GDP.DEFL.KD.ZG` | Inflation / GDP deflator (%) | `API_NY.GDP.DEFL.KD.ZG_DS2_en_excel_v2_121.xls` |

Notes:
- All four files use `skiprows=3` and `sheet_name='Data'`
- The World Bank includes the current year (2025) as an empty placeholder
  column — dropped during cleaning
- Simple filenames are copied from the original downloaded files:
  `cp <original_WB_name>.xls <simple_name>.xls`
- To refresh data: download fresh files from worldbank.org, copy to simple
  names, update the access date above

````python
##
#
# Object Oriented
#
##

class WppPipeline:
    def __init__(self, un_file, wb_gdp_file, wb_gdp_growth_file,
                 wb_gdp_pcap_file, wb_inflation_file,
                 year_start=1961, year_end=2023):
        """
        Class instantiates a pipeline object that loads, cleans, merges,
        normalizes, and builds a time series DataFrame from United Nations
        and World Bank data.

        Parameters
        ----------
        un_file: str, path to UN World Population Prospects Excel file.
        wb_gdp_file: str, path to World Bank GDP Excel file.
        wb_gdp_growth_file: str, path to World Bank GDP growth Excel file.
        wb_gdp_pcap_file: str, path to World Bank GDP per capita Excel file.
        wb_inflation_file: str, path to World Bank inflation Excel file.
        year_start: int, first year to include. Default 1961.
        year_end: int, last year to include. Default 2023.

        Returns
        -------
        self.df: DataFrame, merged and cleaned UN + World Bank dataset.
        self.n_df: DataFrame, normalized version of self.df.
        self.ts_df: DataFrame, normalized time series (OLS coefficients by year).
        """
        self.un_file = un_file
        self.wb_gdp_file = wb_gdp_file
        self.wb_gdp_growth_file = wb_gdp_growth_file
        self.wb_gdp_pcap_file = wb_gdp_pcap_file
        self.wb_inflation_file = wb_inflation_file
        self.year_start = year_start
        self.year_end = year_end

        self.df = None
        self.n_df = None
        self.ts_df = None

    def load_un(self):
        """
        Loads and cleans the UN World Population Prospects Excel file.

        Returns
        -------
        self.df: DataFrame with UN demographic indicators.
        """
        wpp_df = pd.read_excel(self.un_file,
                               sheet_name='Estimates',
                               skiprows=16,
                               thousands=' ',
                               usecols=[
                                   'Index',
                                   'Variant',
                                   'Region, subregion, country or area *',
                                   'ISO3 Alpha-code',
                                   'ISO2 Alpha-code',
                                   'Type',
                                   'Parent code',
                                   'Year',
                                   'Total Population, as of 1 July (thousands)',
                                   'Median Age, as of 1 July (years)',
                                   'Population Growth Rate (percentage)',
                                   'Total Fertility Rate (live births per woman)',
                                   'Life Expectancy at Birth, both sexes (years)',
                                   'Net Number of Migrants (thousands)',
                                   'Net Migration Rate (per 1,000 population)'
                               ])

        wpp_df = wpp_df.rename(columns={
            'Region, subregion, country or area *'         : 'Location',
            'ISO3 Alpha-code'                              : 'ISO3',
            'ISO2 Alpha-code'                              : 'ISO2',
            'Type'                                         : 'LocType',
            'Total Population, as of 1 July (thousands)'  : 'Population_Ks',
            'Median Age, as of 1 July (years)'             : 'MedAge',
            'Population Growth Rate (percentage)'          : 'PopulationGrowthRate',
            'Total Fertility Rate (live births per woman)' : 'FertilityRate_births_per_woman',
            'Life Expectancy at Birth, both sexes (years)' : 'LifeExpectancy',
            'Net Number of Migrants (thousands)'           : 'NetMigrants_Ks',
            'Net Migration Rate (per 1,000 population)'    : 'NetMigrationRate_per_Kpop'
        })

        # Drop rows with no Year (region aggregate header rows)
        wpp_df = wpp_df.dropna(subset=['Year'])
        wpp_df['Year'] = wpp_df['Year'].astype(np.int64)

        # Keep years within range
        wpp_df = wpp_df[(wpp_df['Year'] >= self.year_start) &
                        (wpp_df['Year'] <= self.year_end)]

        # Rename LocType value for consistency
        wpp_df.loc[wpp_df['LocType'] == 'Country/Area', 'LocType'] = 'Country'

        # Add migration direction columns
        wpp_df['ReceivesMigrants'] = [1 if col > 0 else 0
                                      for col in wpp_df['NetMigrants_Ks']]
        wpp_df['ImmigrantsEmigrants'] = ['Immigrants' if col > 0 else 'Emigrants'
                                         for col in wpp_df['NetMigrants_Ks']]

        self._wpp_df = wpp_df
        return self

    def load_wb(self):
        """
        Loads and cleans the four World Bank GDP and inflation Excel files.
        Fills NaN values with column mean, transposes year columns to rows.

        Returns
        -------
        self._wb_df: DataFrame, merged World Bank indicators.
        """
        def _load_and_clean(filepath, indicator, drop_year='2025'):
            df = pd.read_excel(filepath, skiprows=3, sheet_name='Data')
            df = df.drop(drop_year, axis=1)
            df = df.set_index('Country Name')
            df = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
            for column in df:
                df[column] = df[column].replace(np.nan, df[column].mean())
            df = _unpivot(df, 'Location', 'Year', indicator)
            df = df.astype({'Location': str, 'Year': np.int64, indicator: np.float64})
            return df

        def _unpivot(df, var_name, columns_name, indicator):
            N, K = df.shape
            data = {
                indicator    : df.to_numpy().ravel('F'),
                columns_name : np.asarray(df.columns).repeat(N),
                var_name     : np.tile(np.asarray(df.index), K)
            }
            return pd.DataFrame(data, columns=[var_name, columns_name, indicator])

        wb_gdp_df          = _load_and_clean(self.wb_gdp_file,       'GDP_USD')
        wb_gdp_growth_df   = _load_and_clean(self.wb_gdp_growth_file, 'GDP_growth_pct')
        wb_gdp_pcap_df     = _load_and_clean(self.wb_gdp_pcap_file,   'GDP_PCAP_USD')
        wb_inflation_df    = _load_and_clean(self.wb_inflation_file,  'Inflation_pct')

        wb_df = pd.merge(wb_gdp_df,       wb_gdp_growth_df, on=['Location', 'Year'])
        wb_df = pd.merge(wb_df,           wb_gdp_pcap_df,   on=['Location', 'Year'])
        wb_df = pd.merge(wb_df,           wb_inflation_df,  on=['Location', 'Year'])

        self._wb_df = wb_df
        return self

    def merge(self):
        """
        Merges UN and World Bank DataFrames on Location and Year.
        Keeps countries only (removes regions and aggregates).
        Adds log GDP column.

        Returns
        -------
        self.df: DataFrame, merged UN + World Bank dataset, countries only.
        """
        wpp_wb_df = pd.merge(self._wpp_df, self._wb_df, on=['Location', 'Year'])

        # Reorder columns to have migration at the end
        wpp_wb_df = wpp_wb_df[['Index', 'Location', 'ISO3', 'ISO2', 'LocType',
                                'Year', 'Population_Ks', 'MedAge',
                                'PopulationGrowthRate', 'FertilityRate_births_per_woman',
                                'LifeExpectancy', 'ImmigrantsEmigrants', 'ReceivesMigrants',
                                'GDP_USD', 'GDP_growth_pct', 'GDP_PCAP_USD', 'Inflation_pct',
                                'NetMigrants_Ks', 'NetMigrationRate_per_Kpop']]

        # Keep countries only
        wpp_wb_df = wpp_wb_df[wpp_wb_df['LocType'] == 'Country']

        # Add log GDP
        wpp_wb_df['logGDP'] = wpp_wb_df['GDP_USD'].apply(lambda x: log(x) if x > 0 else 0)

        self.df = wpp_wb_df.copy() 
        
        # Ensure numeric columns are float64 — object dtype causes zscore to fail
        self.df['NetMigrationRate_per_Kpop'] = pd.to_numeric(
            self.df['NetMigrationRate_per_Kpop'], errors='coerce')
        self.df['Population_Ks'] = pd.to_numeric(
            self.df['Population_Ks'], errors='coerce')

        # Ensure all expected numeric columns are float64
        numeric_cols = ['Population_Ks', 'MedAge', 'PopulationGrowthRate',
                        'FertilityRate_births_per_woman', 'LifeExpectancy',
                        'NetMigrants_Ks', 'NetMigrationRate_per_Kpop',
                        'GDP_USD', 'GDP_growth_pct', 'GDP_PCAP_USD',
                        'Inflation_pct', 'logGDP']
        for col in numeric_cols:
            self.df[col] = pd.to_numeric(self.df[col], errors='coerce')

        return self

    def remove_outliers(self, z_threshold=2, min_population_ks=1000):
        """
        Removes outlier countries based on z-score of NetMigrationRate_per_Kpop
        and minimum population threshold.

        Parameters
        ----------
        z_threshold: int, z-score threshold for outlier removal. Default 2.
        min_population_ks: int, minimum population in thousands. Default 1000.

        Returns
        -------
        self.df: DataFrame with outlier countries removed.
        self.outlier_countries: list of removed outlier countries.
        self.low_pop_countries: list of removed low population countries.
        """
        outlier_countries = self.df[(np.abs(ss.zscore(
            self.df['NetMigrationRate_per_Kpop'])) > z_threshold)].Location.unique()

        self.df = self.df[~self.df.Location.isin(outlier_countries)]

        low_pop_countries = self.df[
            self.df['Population_Ks'] < min_population_ks].Location.unique()

        self.df = self.df[~self.df.Location.isin(low_pop_countries)]

        self.outlier_countries = outlier_countries
        self.low_pop_countries = low_pop_countries

        return self

    def normalize(self):
        """
        Normalizes selected numeric columns of self.df.
        Normalized columns are prefixed with 'N_'.

        Returns
        -------
        self.n_df: DataFrame, normalized version of self.df.
        """
        cols = ['Population_Ks', 'MedAge', 'PopulationGrowthRate',
                'FertilityRate_births_per_woman', 'LifeExpectancy',
                'GDP_USD', 'logGDP', 'GDP_growth_pct',
                'GDP_PCAP_USD', 'Inflation_pct',
                'NetMigrants_Ks', 'NetMigrationRate_per_Kpop']

        n_df = self.df.copy(deep=True)
        for c in cols:
            if is_numeric_dtype(n_df[c]):
                # Use dropna() to avoid NaN propagation in linalg.norm()
                # Added 2026.03.30
                n_df[c] = n_df[c] / np.linalg.norm(n_df[c].dropna())
                n_df.rename({c: 'N_' + str(c)}, axis=1, inplace=True)

        self.n_df = n_df
        return self

    def build_time_series(self):
        """
        Builds a normalized time series DataFrame using OLS regression.
        For each year, fits OLS models to extract emigration/immigration
        coefficients (B0, B1) as the time series signals.

        Returns
        -------
        self.ts_df: DataFrame with normalized time series, indexed by Year.
        """
        migr_rate_b0, migr_rate_b1 = [], []
        net_migr_b0,  net_migr_b1  = [], []
        gdp_b0,       gdp_b1       = [], []
        log_gdp_b0,   log_gdp_b1   = [], []
        pop_b0,       pop_b1       = [], []
        mean_pop_tot, mean_logGDP, mean_GDP_growth_pct = [], [], []
        year_model = []

        for year in range(self.year_start, self.year_end + 1):

            wpp_tmp = self.n_df[self.n_df['Year'] == year]

            # Drop rows with NaN in OLS columns
            wpp_tmp = wpp_tmp.dropna(subset=['N_GDP_growth_pct',
                                             'N_logGDP',
                                             'N_NetMigrationRate_per_Kpop',
                                             'N_NetMigrants_Ks',
                                             'N_Population_Ks',
                                             'ImmigrantsEmigrants'])

            if len(wpp_tmp) < 2 or wpp_tmp['ImmigrantsEmigrants'].nunique() < 2:
                print(f'WARNING: Skipping year {year} — insufficient data.')
                continue

            # Receive migrants is a categorical variable. Subtract 1 to avoid the intercept.
            ols_mig_rate = smf.ols('N_NetMigrationRate_per_Kpop ~ C(ImmigrantsEmigrants) - 1',
                                   data=wpp_tmp).fit()
            ols_net_migr = smf.ols('N_NetMigrants_Ks ~ C(ImmigrantsEmigrants) - 1',
                                   data=wpp_tmp).fit()
            ols_gdp_grth = smf.ols('N_GDP_growth_pct ~ C(ImmigrantsEmigrants) - 1',
                                   data=wpp_tmp).fit()
            ols_log_gdp  = smf.ols('N_logGDP ~ C(ImmigrantsEmigrants) - 1',
                                   data=wpp_tmp).fit()
            ols_pop      = smf.ols('N_Population_Ks ~ C(ImmigrantsEmigrants) - 1',
                                   data=wpp_tmp).fit()

            migr_rate_b0.append(ols_mig_rate._results.params[0])
            migr_rate_b1.append(ols_mig_rate._results.params[1])
            net_migr_b0.append(ols_net_migr._results.params[0])
            net_migr_b1.append(ols_net_migr._results.params[1])
            gdp_b0.append(ols_gdp_grth._results.params[0])
            gdp_b1.append(ols_gdp_grth._results.params[1])
            log_gdp_b0.append(ols_log_gdp._results.params[0])
            log_gdp_b1.append(ols_log_gdp._results.params[1])
            pop_b0.append(ols_pop._results.params[0])
            pop_b1.append(ols_pop._results.params[1])

            mean_pop_tot.append(wpp_tmp['N_Population_Ks'].mean())
            mean_logGDP.append(wpp_tmp['N_logGDP'].mean())
            mean_GDP_growth_pct.append(wpp_tmp['N_GDP_growth_pct'].mean())
            year_model.append('31DEC' + str(year))

        ts_df = pd.DataFrame({
            'Year'                : year_model,
            'N_EmigrationRate'    : migr_rate_b0,
            'N_ImmigrationRate'   : migr_rate_b1,
            'N_NetEmigration'     : net_migr_b0,
            'N_NetImmigration'    : net_migr_b1,
            'N_EmigrationGDPgrowth'  : gdp_b0,
            'N_ImmigrationGDPgrowth' : gdp_b1,
            'N_EmigrationLogGDP'  : log_gdp_b0,
            'N_ImmigrationLogGDP' : log_gdp_b1,
            'N_EmigrationPop'     : pop_b0,
            'N_ImmigrationPop'    : pop_b1,
            'N_PopulationTot'     : mean_pop_tot,
            'N_logGDP'            : mean_logGDP,
            'N_GDP_growth_pct'    : mean_GDP_growth_pct
        })

        ts_df['Year'] = pd.to_datetime(ts_df['Year'], format='%d%b%Y')
        ts_df.set_index('Year', inplace=True)

        self.ts_df = ts_df
        return self

    def to_csv(self, wpp_wb_path='../data/wpp_wb.csv',
               normalized_path='../data/normalized_wpp_wb.csv',
               ts_path='../data/normalized_world_migration_and_gdp.csv'):
        """
        Writes the three pipeline DataFrames to CSV files.

        Parameters
        ----------
        wpp_wb_path: str, path for merged raw DataFrame. 
        normalized_path: str, path for normalized DataFrame.
        ts_path: str, path for time series DataFrame.
        """
        if self.df is not None:
            self.df.to_csv(wpp_wb_path)
            print(f'Written: {wpp_wb_path}')
        if self.n_df is not None:
            self.n_df.to_csv(normalized_path)
            print(f'Written: {normalized_path}')
        if self.ts_df is not None:
            self.ts_df.to_csv(ts_path)
            print(f'Written: {ts_path}')

        return self

    def build(self):
        """
        Runs the full pipeline in sequence:
        load_un() → load_wb() → merge() → remove_outliers()
        → normalize() → build_time_series()

        Returns
        -------
        self
        """
        return (self
                .load_un()
                .load_wb()
                .merge()
                .remove_outliers()
                .normalize()
                .build_time_series())
````

### _class TimeSeries_

The `TimeSeries` class encapsulates all time series operations used in this 
study. Instantiating it with a pandas Series and an optional `train_size` 
parameter automatically creates train and test splits.

**Methods:**
- `lineplots_ts()` — line plots for each numeric column
- `plot_hist_qq()` — histogram and Q-Q plot for normality assessment
- `acf_pacf_plots()` — ACF and PACF plots
- `test_ADF_stationary()` — Augmented Dickey-Fuller stationarity test
- `decompose()` — seasonal decomposition (trend, seasonal, residual)
- `arima()` — fit an ARIMA / SARIMAX model
- `auto_arima()` — auto-select best ARIMA model via BIC
- `test_markov_regime()` — Markov switching regression regime test
- `get_train_test()` — return train and test splits
- `predict()` — forecast using auto_arima on train/test split
- `predict_plot()` — plot forecasts with confidence intervals

**Data available after instantiation:**
- `ts.ts` — the original time series passed in
- `ts.train` — training split (if `train_size > 0`)
- `ts.test` — test split (if `train_size > 0`)
- `ts.train_size` — number of observations in the training set

**Usage:**
```python

# Or load from CSV as before
```

````python
##
#
# Objected Oriented
#
##

class TimeSeries:
    def __init__(self, ts, train_size=0):
        """
        Class instantiates object that contains a time series.
        
        Parameters
        ----------
        ts: 2D array or DataFrame with index type datatime and one numeric columns.
        train_size: int, to create train and test sets.
        
        Returns
        -------
        tain, test: time series train and test
        """
        #your code here
        self.ts = ts
        self.train_size = train_size
        
        if train_size > 0:
            # needs from pmdarima import model_selection
            self.train, self.test = model_selection.train_test_split(self.ts, train_size=self.train_size)
            
    def lineplots_ts(self):
        """
        The plots aim to display time series in general.
        Each numeric column of the df is expected to be a time series.
        It plots each numeric column.
        """
        no_of_plots = self.ts.shape[1]
        fig, ax = plt.subplots(self.ts.shape[1], 1, figsize=(10, 2*no_of_plots))
        ax = ax.flatten()

        for i in range(self.ts.shape[1]):
            if is_numeric_dtype(self.ts[self.ts.columns[i]]):
                sns.lineplot(x = self.ts.index, y = self.ts.columns[i], data = self.ts, ax=ax[i])
                ax[i].axhline(y = self.ts[self.ts.columns[i]].mean(), color = 'r', linestyle = '-')

                    
    def plot_hist_qq(self, title='time series', bins=40):
        """
        Plots histogram and Q-Q to visualize distribution and Normality.
        
        Parameters
        ----------
        title: string with title for the histogram plot.
        bins: int with number of bins to plot the histogram.
        """
        # plot_hist_and_qq(sel.ts, title=title, bins=bins)
        
        _,ax = plt.subplots(1,2,figsize=(10,4))
        
        ## Plot the histogram with labels
        self.ts.hist(ax = ax[0], bins = bins, alpha = 0.5);
        ax[0].set_xlabel('Value');
        ax[0].set_ylabel('Frequency');
        ax[0].set_title('Histogram of ' + title);
        ## Plot the q-q plot on the other axes
        ss.probplot(self.ts, plot = ax[1]);

        
    def acf_pacf_plots(self):
        """
        Plots Auto Correlation Function (ACF) and Partial ACF (PACF).
        """
        lags = len(self.ts) // 3
        print('lags = ', lags)
        _,ax = plt.subplots(2,1, figsize=(8,8))
        _=splt.plot_acf(self.ts, lags = lags, ax=ax[0]);
        _=splt.plot_pacf(self.ts, lags = lags, method='yw', ax=ax[1]);
        
    def test_ADF_stationary(self):
        """
        Augmented Dickey-Fuller (ADF) method to test stationary.
        
        Returns:
        ADF_dict: dictionary with results. ADF static, p-value, lags, key_values
    
        p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.
        p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.
        https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html
        https://www.kaggle.com/code/iamleonie/time-series-interpreting-acf-and-pacf
        https://machinelearningmastery.com/time-series-data-stationary-python/
        https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html
        """
        stationary = adfuller(self.ts)
        ADF_dict = {'test_stat':stationary[0],
                    'p-value':stationary[1],
                    'lags':stationary[2],
                    't-value_10%_conf':stationary[4]['10%'],
                    't-value_5%_conf':stationary[4]['5%'],
                    't-value_1%_conf':stationary[4]['1%']
                   }

        ## Print the results
        print('ADF test statistic = ' + str(ADF_dict['test_stat']))
        print('p-value = ' + str(ADF_dict['p-value']))
        print('Critical Values:')
        # stationary[4] is a dictionary {'10%':float, 1%':float, '5%':float}
        for key, value in stationary[4].items():
            print('\t%s: %.3f' % (key, value))

        # To reject the H0 null hypothesis, the p-value needs to be < 0.05 (rule of thumb) and
        # the t-value at confidence interval 5% needs to be larger than the ADF test static (critical value)
        # Note that larger is not abs(), so more negative is smaller than less negative
        if (ADF_dict['p-value'] <= 0.05) & (ADF_dict['t-value_5%_conf'] > ADF_dict['test_stat']):
            print("*** Time series is stationary.")
            # The series does not have a unit root and it is stationary.
            print("*** Reject the null hypothesis (H0). Both p-value <= 0.05, and t-test 5% conf interval > test static.")
        else:
            print("*** Time series is non-stationary.")
            print("*** Failed to reject the null hypothesis (H0). Either p-value > 0.05, or t-test 5% conf interval < test static.")
    
        print('Number of lags used = ' + str(stationary[2]))
    
        return ADF_dict
    
    def decompose(self, extrapolate_trend='freq', period=12):
        """
        Decomposes the time series into residual, trend, and seasonal series, and plots them.
        
        Parameters
        ----------
        extrtapolate_trend: str, 
        period: int, number of periods, depending on time series. May be in days, month, years, etc.
        
        Returns:
        DataFrame: time series datetime index, columns 'resid', 'trend', 'seasonal'.
        """
        res = sts.seasonal_decompose(self.ts, extrapolate_trend = extrapolate_trend, period = period) 
        res.plot()
        return(pd.DataFrame({'resid': res.resid, 
                             'trend': res.trend, 
                             'seasonal': res.seasonal}, 
                            index = self.ts.index))

    def arima(self, exog=None, order=(0,0,0), seasonal_order=(0, 0, 0, 0), freq=None):
        """
        Return ARIMA model fit.

        Parameters
        ----------
        exog: array
        order: tuple with three components p, d, q
        seasons_order: tuple with four component P, D, Q, S
        freq: str, 'A-DEC' as example

        Returns: 
        ARIMA model fit()
        """

        # NOTE (2026-03-26): freq= parameter was removed in statsmodels 0.14.
        # To update in Phase 2: remove freq=freq from ARIMA() call.
        model = ARIMA(self.ts, exog=exog, order=order, seasonal_order=seasonal_order, freq=freq)
        model_fit = model.fit()
        print('freq used = ', freq)
        print(model_fit.summary())
        return(model_fit)
       
    def auto_arima(self, exogenous, start_p=1, start_q=1, max_p=3, max_q=3, m=10,
               start_P=0, seasonal=True, d=1, D=1, trace=True, information_criterion = 'bic',
               stepwise=True):
        """
        Runs multiple ARIMA models and selects bets model.

        Parameters
        ----------
        exogenous: 2D array or DataFrame with index type datatime and one numeric columns.
        start_p: default 1 
        start_q: defaults 1 
        max_p: default 3
        max_q: default 3 
        m: int, number of periods <-- important
        start_P: default 0 
        seasonal: Logical True/Fals, default True 
        d: default 1
        D: default 1
        trace: logical True/False default True
        information_criterion: str, default 'bic'
        stepwis: Logical True/False default True

        Returns: 
        ARIMA model fit()
        """
        auto_model = pm.auto_arima(self.ts, exogenous=exogenous, start_p=start_p, start_q=start_q,
                                   max_p=max_p, max_q=max_q, m=m, start_P=start_P, seasonal=seasonal,
                                   d=d, D=D, trace=trace, information_criterion=information_criterion,
                                   stepwise=stepwise)
        return auto_model

    def test_markov_regime(self, exog=None, k_regimes=2, freq='YE'):
        """
        'Markov switching models in statsmodels to estimate dynamic regression models with changes in regime.'
        Calculates Markov regression probabilities of regine changes, plots the series, and returns the series.

        Parameters
        ----------
        endog: numeric series
        exog: numeric series
        k_regimes: int, number of regines
        freq: sdtr, 'Y' default

        Returns: 
        --------
        the model fit
        
        References:
        -----------
        https://www.statsmodels.org/stable/examples/notebooks/generated/markov_regression.html
        http://www.stata.com/manuals14/tsmswitch.pdf
        
        sm.tsa.MarkovRegression.__init__
        sm.tsa.MarkovRegression.__dict__

        """
        # NOTE (2026-03-26): freq= parameter is deprecated in MarkovRegression().
        # To update in Phase 2: remove freq=freq from MarkovRegression() call.
        markov_regression_fit = sm.tsa.MarkovRegression(endog=self.ts, exog=exog, k_regimes=k_regimes, freq=freq).fit()
        markov_regression_probs = markov_regression_fit.smoothed_marginal_probabilities[1]
        markov_regression_probs.plot(title="Markov Regression Regime Test")
        
        return markov_regression_fit

    def get_train_test(self):
        """
        Obtain train and test samples from the list
         
        Parameter:
        ----------
        train_size: int
        
        Returns:
        --------
        train, test lists
        """
        
        train = self.train
        test = self.test
    
        return train, test

    def predict(self, train_exog=None, test_exog=None, trace=True, m=5, time_start=1960):
        """
        Predict and plot based on auto_arima model and on a train/test splif of the time series
        
        Parameters:
        -----------
        train: time series
        test: time series
        train_exog: time series
        test_exog: time series
        
        Returns:
        --------
        preds, conf_int: predictions and confidence intervals
        """
        
        train = self.train
        test = self.test
    
        mod_auto_arima = pm.auto_arima(train, exogenous=train_exog, start_p=1, start_q=1, start_P=1, start_Q=1,
                         max_p=3, max_q=3, max_P=3, max_Q=5, seasonal=True, trace=trace, m=m,
                         information_criterion = 'bic', stepwise=True, suppress_warnings=True, D=1,
                         error_action='ignore')

        # The script is based on example provided by alkaline-ml.com:
        # http://alkaline-ml.com/pmdarima/1.0.0/modules/generated/pmdarima.arima.auto_arima.html
        # http://alkaline-ml.com/pmdarima/1.0.0/auto_examples/arima/example_auto_arima.html#sphx-glr-auto-examples-arima-example-auto-arima-py
        # Additional reference:
        # https://en.wikipedia.org/wiki/Autoregressive_integrated_moving_average

        preds, conf_int = mod_auto_arima.predict(n_periods=test.shape[0], exogenous=test_exog, return_conf_int=True)

        # Print the error:
        print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(test, preds)))

        return preds, conf_int

    def predict_plot(self, preds, conf_int, time_start=1960, title='Forecasts', xlabel='Year'):
        """
        Predict and plot based on auto_arima model.
        
        Parameters:
        -----------
        train: time series
        test: time series
        preds: time series, predictons
        test_exog: time series
        
        Returns:
        --------
        None
        """

        train = self.train
        test = self.test
        
        # This script is based and takes code from the example provided in:
        # http://alkaline-ml.com/pmdarima/1.0.0/modules/generated/pmdarima.arima.auto_arima.html
        # http://alkaline-ml.com/pmdarima/1.0.0/auto_examples/arima/example_auto_arima.html#sphx-glr-auto-examples-arima-example-auto-arima-py
        
        # Plot the points and the forecasts
        x_axis = np.arange(train.shape[0] + preds.shape[0])
        x_years = x_axis + time_start  # For example, year starts at 1960

        plt.plot(x_years[x_axis[:train.shape[0]]], train, alpha=0.75)
        plt.plot(x_years[x_axis[train.shape[0]:]], preds, alpha=0.75)  # Forecasts
        plt.scatter(x_years[x_axis[train.shape[0]:]], test, alpha=0.4, marker='x')  # Test data
        plt.fill_between(x_years[x_axis[-preds.shape[0]:]], conf_int[:, 0], conf_int[:, 1], alpha=0.1, color='b')
        plt.title(title)
        plt.xlabel(xlabel)
        
        return None
````

## Time Series Analysis Worldwide View

These next section will explor **Seasonal, Trend and Residual (STL)** models, consisting of:

- Trend removals with a LOESS regression model. 
- Seasonal component removals using regression on periodic components.
- Leaving a remainder known as the Residual. 

This study applies an **additive** approach, $TS(t) = S(t) + T(t) + R(t)$.

Reference: https://www.statsmodels.org/stable/generated/statsmodels.tsa.seasonal.seasonal_decompose.html


Important parameters:
- **p** Auto Regressive (AR) order.
- **d** number of differences.
- **q** order of Moving Average (MA).
- **P** seasonal AR term order. Autoregressive component with period **M**.  
- **D** order of differencing for non-stationary behavior. 
- **Q** seasonal MA term order. Mving average component with period **M**. 
- **S** period length.
