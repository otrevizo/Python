# OpenAI Hello World

```python
import os                         # To read environment variables
from openai import OpenAI         # Modern OpenAI client (v1.0+)
import anthropic                  # Anthropic Claude client
from datetime import date

today = date.today().strftime("%B %d")
```

```python
import openai
import anthropic as anth
print(f'OpenAI version:    {openai.__version__}')
print(f'Anthropic version: {anth.__version__}')
```

## Instantiate the OpenAI client

The modern pattern (openai >= 1.0) uses a client object instead of
setting `openai.api_key` globally. The key is read from the environment.


```python
# Instantiate the OpenAI client — reads OPENAI_API_KEY from environment
openai_client = OpenAI(api_key=os.environ['OPENAI_API_KEY'])
```

## Send a message and print the response

Using `gpt-4o-mini` — the current recommended model for experimentation.
Replaces the deprecated `gpt-3.5-turbo`.


```python
completion = openai_client.chat.completions.create(
    # https://platform.openai.com/docs/models
    model='gpt-4o-mini',          # gpt-3.5-turbo deprecated late 2024
    messages=[
        {
            'role': 'user',
            'content': f'Hello, world! Today is {today}. What happened on this day in history?',
        },
    ],
)

# Print the response content
print(completion.choices[0].message.content)
```

# Anthropic Claude Hello World

Same pattern as OpenAI — instantiate a client, send a message, print the response.
The Anthropic API uses `messages.create()` with a `max_tokens` parameter.


## Instantiate the Anthropic client


```python
# Instantiate the Anthropic client — reads ANTHROPIC_API_KEY from environment
anthropic_client = anthropic.Anthropic(api_key=os.environ['ANTHROPIC_API_KEY'])
```

## Send a message and print the response

Using `claude-sonnet-4-5` — fast, capable, and economical. The right default
for experimentation, analogous to `gpt-4o-mini` on the OpenAI side.


```python
from datetime import date
today = date.today().strftime("%B %d")

message = anthropic_client.messages.create(
    model='claude-sonnet-4-5',
    max_tokens=1024,
    messages=[
        {
            'role': 'user',
            'content': f'Hello, world! Today is {today}. What happened on this day in history?',
        },
    ],
)
print(message.content[0].text)
```

## Pattern Comparison

| | OpenAI | Anthropic |
|---|---|---|
| **Library** | `openai` | `anthropic` |
| **Client** | `OpenAI()` | `Anthropic()` |
| **Env var** | `OPENAI_API_KEY` | `ANTHROPIC_API_KEY` |
| **Method** | `client.chat.completions.create()` | `client.messages.create()` |
| **Model** | `gpt-4o-mini` | `claude-sonnet-4-5` |
| **max_tokens** | optional | required |
| **Response** | `completion.choices[0].message.content` | `message.content[0].text` |
