```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#####
#
# Sigmoid
#
####

def my_sigmoid(b = 0, theta = 0.1, shift = 500, space = 1000, exodous = None, plot = False):
    """
    Reversed sigmoid with some shif and damping effect
    
    Parameters:
    -----------
    b: int the start of the function, normally 0, but it can be moved
    theta: float growth factor, normally 0.1... A negative theta reverses the sigmoid
    shift: int, a number where the sigmoid will cut, traverse
    space = int, number of measurements
    exodous: dataframe, optional, reference to be used to extract its length if necessary
    
    Returns:
    --------
    r: NumPy array with the reverse sigmoid function
    """

    if exodous != None: space = len(exodous) - b
    x = np.linspace(b, space, space)
    growth = 1
    r = np.exp(theta*(x - shift)) / (1 + np.exp(theta*(x - shift) ))
    
    if plot == True:
        plt.plot(x, r)
        
    return r
```

```python
r = my_sigmoid(b = 0, theta = 0.05, plot = True)
```

```python
r = my_sigmoid(b = 0, theta = -0.05, plot = True)
```
