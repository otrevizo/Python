```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#####
#
# Sigmoid
#
####

def my_sigmoid(start = 0, stop = 1000, num = 1000, theta = 0.1, shift = 500, plot = False):
    """
    Reversed sigmoid with some shif and damping effect
    
    Parameters:
    -----------
    start: int the start of the function, normally 0, but it can be moved
    stop = int, for linespace
    num: int, number of points
    theta: float growth factor, normally 0.1... A negative theta reverses the sigmoid
    shift: int, a number where the sigmoid will cut, traverse
    
    Returns:
    --------
    r: NumPy array with the reverse sigmoid function
    """
    x = np.linspace(start, stop, num)
    r = np.exp(theta*(x - shift)) / (1 + np.exp(theta*(x - shift) ))    
    if plot == True: plt.plot(x, r)        
    return r
```

```python
r = my_sigmoid(start = 0, theta = 0.05, plot = True)
```

```python
r = my_sigmoid(start = 0, theta = -0.05, plot = True)
```

```python
r = my_sigmoid(start = 200, stop = 2000, num = 1800, theta = 0.05, shift = 700, plot = True)
```

```python
r = my_sigmoid(start = 200, stop = 2000, num = 1800, theta = -0.05, shift = 700, plot = True)
```
