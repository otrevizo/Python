# Descriptive Statistics

### _Mean_:
$$\mu = \frac{\sum_{i=1}^Nx_i}{N}$$

For a sample, the mean is normally depicted by x-bar: $\bar x$

### _Standard Deviation_:
Population _standard deviation_: 
$$\sigma=\sqrt{\frac{\sum_{i=1}^N (x_i-\mu)^2}{N}}$$

Sample _standard deviation_: 
$$\sigma_s=\sqrt{\frac{\sum_{i=1}^N (x_i-\bar x)^2}{N-1}}$$

### _Variance_:
$$variance = \sigma^2$$

$$variance_s = \sigma_s^2$$

### _Z-score_:
_Z-Score_ gives you the number of s.d. from the mean.

I.e., the _Z-score_ normalizes the number of standard daviations away from the mean: 

$$Z_x=\frac{(x-\mu)}{\sigma}$$

$$Z_{x_s}=\frac{(x-\bar x)}{\sigma_s}$$

_Z-distribution_ has $\sigma=1$ and $\mu = 1$.

(a) If $\sigma$ is available, use the _Z-distribution_. Otherwise, use $\sigma_s$ and _t-distribution_.

(b) If the $sample \ge 30$, use the _Z-distribution_. Otherwise, use $\sigma_s$ and _t-distribution_.

# Inferential Statistics


Linear Regression aims to model a *predictor* and *response* association based on a straight line. The *response* $Y$ is predicted by an equation that contains coeffcients $\beta_0$ and $\beta_1$ plus a residual error $e$ as stated in the following equation.

$Y=\beta_0+\beta_1X+e$

The goal of *linear regression* is to estimate coefficients $\beta_0$ and $\beta_1$ and evaluate **two hypotheses**: A a **null hypothesis** $H_0$ and an **alternative hypothesis** $H_a$ defined as follows:

$H_0$: There is no relationship between $X$ and $Y$

$H_a$: There is some relationship between $X$ and $Y$

$H_0$: $\beta_1 = 0$

versus

$H_a$: $\beta_1 \neq 0$

The null hypothesis states; *there is no relationship between the two variables.* The alternative hypothesis states; _there is some relationship between the two variables._ A very small *p-value* (in general *p-value* $<0.05$) indicates that the null hypotheses $H_0$ is not plausible.

Therefore, when we have very a small *p-value* we can *reject* the null hypothesis. That allows us to infer that _"there is an association between the predictor and the response"_ (ISLR p.68).

And the results from `summary(lm(...))` shows *three asteriscks* next to the *p-value* indicating that the relationship is statistically **significant.** One would need to do additional tests on an additional data samples if possible to make sure the conclusion is not tied to this specific dataset. Therefore, **there is statistically significant association between these two variables, `contrib` and `lastcontr`,** based on this summary.



# Probability

### Bernoulli distributions

_(Source of Bernoulli distr. section, Prof. Elston, Harvard University)._

Bernoulli distributions model the outcome of a **single trial** with two possible outcome: success or failure, a binary outcome that can be quantified as $0 or 1$. 

There is a probability $p$ of success. The probability mass function for the Bernoulli distribution as:

\begin{align}
x \in {0, 1}\\
P(x\ |\ p) &= \bigg\{ 
\begin{matrix}
p\ if\ x = 1\\
(1 - p)\ if\ x = 0
\end{matrix}\\
or\\
P(x\ |\ p) &= p^x(1 - p)^{(1-x)}\\ 
\end{align}

\begin{align}
Mean &= p\\
Variance &= p(1-p)
\end{align}

### Binomial distribution

A Bernoulli trial has two possible outcomes: success or failure. A binomial distribution applie to a number k of successful outcomes out of N trials.

The probability of k successful outcomes out of N trials is:

$$P(k\ |\ N, p) = \binom{N}{k} p^k(1 - p)^{(N-k)}$$


Where the binomial coefficient is defined as _"N choose k"_: $$\binom{N}{k} = \frac{N!}{k!(N-k)!}$$

Bernoulli mean and variance:

$$Mean = Np$$
$$Variance = Np(1-p)$$


### Poisson Distribution    

_(Source of Bernoulli distr. section, Prof. Elston, Harvard University)._

"The Poisson distribution models the occurrence of events in a fixed interval of time. We say that the Poisson distribution models the probability, $P$, of x **arrivals** within the time period."

Poisson distribution average arrival rate is $\lambda$:

$$ 
P(x\ |\ \lambda) = \frac{\lambda^x}{x!} \exp^{-\lambda}
$$

The mean and variance of the Poisson distribution are both equal to the parameter $\lambda$, or:

\begin{align}
Mean &= \lambda\\
Variance &= \lambda
\end{align}
